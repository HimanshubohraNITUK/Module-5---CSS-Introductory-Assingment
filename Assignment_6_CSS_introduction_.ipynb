```python
html code
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>CSS Priority Example</title>

  <!--  Linking External CSS -->
  <link rel="stylesheet" href="style.css" />

  <!--  Internal CSS: Will override external if there's conflict -->
  <style>
    .box {
      background-color: lightblue;
      color: black;
      padding: 20px;
      margin: 10px;
      border: 2px solid blue;
    }
  </style>
</head>
<body>

  <!-- Inline CSS: Has the highest priority -->
  <div class="box" style="background-color: orange; color: white;">
    This div uses Inline + Internal + External CSS.
  </div>

  <div class="box">
    This div uses Internal + External CSS.
  </div>

  <div class="only-external">
    This div uses only External CSS.
  </div>

  <!--  Notes:
    - External CSS comes from styles.css
    - Internal CSS is in <style> tag in head
    - Inline CSS is added directly in the HTML element
    - Priority: Inline > Internal > External
  -->

</body>
</html>

css code

/* External CSS (lowest priority if overridden) */
.box {
  background-color: lightgreen;
  color: darkgreen;
  border: 2px dashed green;
}

.only-external {
  background-color: lightgray;
  color: black;
  padding: 20px;
  margin: 10px;
  border: 2px solid gray;
}
```

Q2 - Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.


```python
-
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>BEM Paragraph Styling</title>

  <style>
    /* Base Block */
    .content__paragraph {
      font-family: Arial, sans-serif;
      padding: 10px;
      margin-bottom: 15px;
      border-left: 5px solid transparent;
    }

    /* Modifier 1: Red highlighted */
    .content__paragraph--important {
      background-color: #ffe5e5;
      color: #cc0000;
      border-left-color: #cc0000;
    }

    /* Modifier 2: Info style */
    .content__paragraph--info {
      background-color: #e5f4ff;
      color: #005b99;
      border-left-color: #005b99;
    }

    /* Modifier 3: Success style */
    .content__paragraph--success {
      background-color: #e6ffea;
      color: #2e7d32;
      border-left-color: #2e7d32;
    }

    /* Modifier 4: Warning style */
    .content__paragraph--warning {
      background-color: #fff8e1;
      color: #ff6f00;
      border-left-color: #ff6f00;
    }
  </style>
</head>
<body>

  <!-- Default style (base class only) -->
  <p class="content__paragraph">
    This is a normal paragraph using the base BEM block style.
  </p>

  <!-- Important Modifier -->
  <p class="content__paragraph content__paragraph--important">
    This is an important message. Styled using the '--important' modifier.
  </p>

  <!-- Info Modifier -->
  <p class="content__paragraph content__paragraph--info">
    This is an informational paragraph with calm blue tones.
  </p>

  <!-- Success Modifier -->
  <p class="content__paragraph content__paragraph--success">
    This paragraph shows a success message.
  </p>

  <!-- Warning Modifier -->
  <p class="content__paragraph content__paragraph--warning">
    This paragraph contains a warning message.
  </p>

</body>
</html>
```
